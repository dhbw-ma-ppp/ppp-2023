```python
import pandas as pd
import seaborn as sns
from pathlib import Path
titanic_data = Path.cwd().parent.parent / 'data' / 'titanic.csv'
df_sex_age = pd.read_csv(titanic_data)
```

```python
sns.displot(data=df_sex_age, col="Sex",x="Age",hue="Survived")
```

this plot shows the count of dead and alive passengers seperated by Sex. 
You can see that the count of dead men is way above the alive count and for women it is the other way around.

```python
sns.displot(data=df_sex_age, col= "Survived", x ="Age", hue="Sex", y = "Pclass")
```

this plot shows the Age and Sex of dead and surviving passengers based on their class.
You can see that people who died were mostly male. In the third class the survival chance is about 50% for women, which makes it the most unsafe class.
male passengers in the middle age are most likely to die.
male passengers in general are way more likely to die than female passengers.
