```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

file_path = "C:/Users/Paul/Documents/Uni/Python/Python Kurs/data/titanic.csv"

tab = pd.read_csv(file_path, delimiter=",")

tab["New_age"] = tab["Age"].apply(lambda x: (x//10)*10)

tab.groupby("New_age").count().plot.bar(y="Survived")
```

Stat1:

This statistic shows, how many passengers there were on the Titanic grouped by ages 0-9, 10-19, 20-29, ...

This helps to understand the dataset better, but isn't relevant for later analysis.

As you can see, most passengers were in the age-groups 20-29 and 30-39. Around that, the distributions look like a Gaussian Curve (Bell Curve), as it should be according to mathematical probability theory.

```python
tab.groupby("New_age").sum("Survived").plot.bar(y="Survived")
```

Stat2:

Above you can see the total number of survivors grouped by the age-groups 0-10, 11-20, 21-30, ...

This information is useful for better understanding of the following statistic, but not relevant for later use.

As you can see, the form's curve looks pretty much the same as the curve above. The most noticeable difference is that the very young (0-9 year olds) survived more likely

```python
tab.groupby("New_age").mean("Survived").plot.bar(y="Survived")
```

Stat3:

This statistic shows the percentage of survivors in the age-groups. This helps to analyze, if the age of a passenger had an influence on his/her survival.

The conclusion of the last diagram fits to this statistic. You can see, that the bar of the 0-9 year olds is the highest (if you ignore the 80-89 year olds)

To understand, why the bar of the 80+ year olds is at 100%, it's important to know, how many 80+ year olds there were in total:

```python
age_groups = tab.groupby("New_age").groups
len(age_groups[80.0])
```

The table contains only one person, who's in the age-group 80-89. Because this one person survived, the bar in Stat3 is at 100%, which can be neglected. One example is just not sufficient for a statistical statement.

```python
tab.groupby("Sex").mean("Survived").plot.bar(y="Survived")
```

Stat4:

This statistic shows the percentage of survivors grouped by sex.

Here you can see, that more women survived, than men. This confirms the use of the "Ladies first" tactic in evacuation, as it is shown in the movie.

```python
tab.groupby("Pclass").mean("Survived").plot.bar(y="Survived")
```

Stat5:

This statistic shows the percentage of survivors grouped by the passenger class.

As you can see, the higher the class, the higher a persons chance of survival. It seems like, the evacuation of the better paying customers had more priority?

To underline my suspision, I will analyze the death-rate depending on the amount of money, the person spend for their ticket.

```python
tab["Fare_group"] = tab["Fare"].apply(lambda x: (x//10)*10)

tab.groupby("Fare_group").mean("Survived").plot.bar(y="Survived")
```

Stat6:

The trend clearly shows, that the people, who are paying more statistically survived more likely. That underlines my statement from Stat5

Summary:
In Conclusion, the age, sex and the amount of paid money influenced the chance of survival of the passenger.
