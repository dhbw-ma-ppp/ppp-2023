## Importing libraries 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Importing data

```python
file_path = "./../../data/titanic.csv"
```

## 1.General Data Exploration and Data Analysis

```python
df = pd.read_csv(file_path, delimiter=",")

df['Age'].plot(kind='hist')
```

The majority of individuals aboard the ship fell within the age range of 20 to 35.
A noticeable decline is observed prior to reaching 20 years old, while a more gradual decrease
is evident beyond 35 years old.

```python
sex_distribution = df['Sex'].value_counts(normalize=True) * 100
sex_distribution.plot(kind='bar', rot=0)
plt.title('Percentage Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.show()
```

The number of male passengers outnumbered that of female passengers on the ship.

```python
df.groupby("Pclass").mean("Survived").plot.bar(y="Survived")
```

As anticipated, individuals in higher classes exhibited a greater survival rate. This may be attributed to prioritization during rescue efforts or the proximity of cabins belonging to those in higher classes to the lifeboats.

```python
df['Embarked'].value_counts().plot(kind='bar')
```

The majority of individuals on the ship boarded from Southampton, with Cherbourg as the second most common departure point, followed by Queenstown.
