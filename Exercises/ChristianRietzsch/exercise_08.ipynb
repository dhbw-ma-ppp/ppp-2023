# The Dataset
## Things that can be analysed:
survival-rate based on gender, class, parch, family-members, cabin, location

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
data = pd.read_csv("../../data/titanic.csv")
data
```

```python
age_survived = data.groupby('Age')['Survived']

age_survived.sum('Survived').plot(color="blue")
leg = age_survived.count().plot(color="red", title="Number of Survivors by Age")
leg.legend(["Survivors", "All People"])
```

The red graph shows all people based on their age.  
The blue graph shows how many people of a certain age survived.  
You can see that most people were between 15 and 35 years old.  
Therefore the number of survivors in this age bracket is the highest.  

```python
data.groupby('Sex')['Survived'].mean('Survived').plot.bar(title="Survival-Rate based on gender")
```

This graph shows the survival rate (in percent) based on gender.  
You can see that women were far more likely to survive 

```python
data.groupby('Pclass')['Survived'].mean('Survived').plot.bar(title="Survival-Rate based on class")
```

This shows the survival rate based on the class a person was residing in  
As expected, people from the first class had the highest chance to survive

```python
data.groupby('Embarked')['Survived'].mean('Survived').plot.bar(title = "Survival-Rate based on location")
```

```python
data.groupby('Parch')['Survived'].mean('Survived').plot.bar(title = "Survival-Rate based on the number of children per person")
```

```python
data.groupby('SibSp')['Survived'].sum('Survived').plot()
leg = data.groupby('SibSp')['Survived'].count().plot(title="Survival-Rate based on the number of family members on the ship")
leg.legend(["Survivors","Alle People"])
```
