```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Data/titanic.csv')
#df
```

```python
sns.barplot(
    data=df,
    x = "Fare",
    y = "Sex",
)
```

In this graph we can see that women paid a higher Fare than men. This is caused by more men being in the third class and therefore paying the least fare

```python
sns.relplot(
    data=df,
    x = "Age",
    y = "SibSp",
    hue = "Survived",
    col = "Sex")
```

This graph shows us that there were more boys/men with a higher amount of siblings onboard which did not survive. Women/girls with low amount of siblings had the highest amount of survivors

```python
sns.jointplot(data=df, 
                x="Age", 
                y="Fare", 
                hue="Sex",
                #col="Survived",
                kind="kde"
             )


```

```python
sns.jointplot(data=df, 
                x="Age", 
                y="Pclass", 
                hue="Survived",
                #col="Survived",
                kind="scatter"
)
```

```python
sns.displot(data=df, x="Age", col="Pclass", kde=True);
sns.displot(data=df, x="Age", col="Survived", kde=True);
sns.displot(data=df, x="Sex", col="Pclass", kde=False);
```

Men dominated the third class while the first class and second class was not dominated by each of both genders. Most men had an age of about 25years in third class

```python
sns.relplot(
    data=df,
    x = "Age",
    y = "Cabin",
    hue = "Survived",
    col="Pclass", style = "Sex")
```

# What does this mean now?
Yet, we see that women paid more money for their titanic-adventure. There were much more women who survived that were older than 10 years. The passengers of first class had a much higher rate of survival. People in third class had the least chance of survival. The third class was dominated by men in the age of around 25. People in first class below the age of 60 had the highest chance of survival. Especially children had a really high chance of suvival just like women.
